crawl DOI for the later usage to download semantic scholar papers by DOI

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

from socket import error as socket_error
import errno


headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/605.1.15'
                         '(KHTML, like Gecko) Version/12.1.1 Safari/605.1.15'}
```

```python
def crawl_DOI(url):
    
    try:
        r = requests.get(url, headers=headers)
    except requests.exceptions.ConnectionError:
        print('Connection refused!')
    except socket_error as e:
        if e.errno != errno.ECONNRESET:
            raise
        pass
    
    try:
        html = r.text
    except:
        html = ""

    # extract the DOI
    try:
        title = re.findall('title="To the current version of the paper by DOI">(.*?)</a>', html)[0]
    except:
        title = ""

    return title
```

```python
df['DOI'] = df['url'].apply(crawl_DOI)
```
